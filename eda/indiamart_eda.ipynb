### Importing Pandas and MatPlotLib Libraries and Reading the Data stored after ETL Process

```python
import pandas as pd
import matplotlib.pyplot as plot

indiamart_df = pd.read_csv("../data/indiamart_flyash_data.csv")

indiamart_df.head()
```

### Basic Data Understanding of Types and Other Attributes

```python
indiamart_df.info()
indiamart_df.describe()
```

### Checking for any missing values in the DataFrame

```python
missing_values = (indiamart_df.isna().mean().sort_values(ascending=False).mul(100))
missing_values
```

### Name of Unique Suppliers

```python
unique_supplier = indiamart_df["supplier_name"].unique()
unique_supplier
```

### Names of Unique Products

```python
unique_products = indiamart_df["product_name"].unique()
unique_products
```

### Top 10 Supplier

```python
top_suppliers = indiamart_df["supplier_name"].value_counts().head(10)
top_suppliers
```

### Plotting of bar graph of Top 10 Unique Suppliers by Produc Count

```python
top_suppliers.plot(kind="bar", figsize=(10,4))
plot.title("Top 10 Suppliers by Product Count")
plot.ylabel("Number of Products")
plot.xlabel("Supplier")
plot.show()
```

### Top 10 States 

```python
top_states = indiamart_df["state"].value_counts().head(10)
top_states
```

### Plotting of bar graph of Top 10 States by No of Products supplied

```python
top_states.plot(kind="bar", figsize=(10,4))
plot.title("Top states suppliers of Products")
plot.xlabel("States Name")
plot.ylabel("No of Products supplied")
plot.show()
```

### Price Distribution of Fly Ash Bricks Machine and Plotting the Graph

```python
prices = indiamart_df["price_in_rupees"].str.replace(",", "", regex=True).astype(int)
prices.plot(kind="hist", bins=30, figsize=(8, 4))

plot.title("Price Distribution of Fly Ash Bricks Machine")
plot.xlabel("Price (INR)")
plot.ylabel("Count")
plot.show()
```

### Value of Product and their ratings graph

```python
indiamart_df.plot(x="price_in_rupees", y="rating", kind="scatter", figsize=(10,7))

plot.title("Price vs Rating")
plot.xlabel("Price(INR)")
plot.ylabel("Rating")
plot.show()
```

### Rating of different products by Product ID

```python
indiamart_df["product_id"] = indiamart_df["product_id"].astype(str)

indiamart_df.plot(
    x="product_id",
    y="rating",
    kind="scatter",
    figsize=(20, 4)
)
plot.title("Product ID vs Rating")
plot.xlabel("Product ID")
plot.ylabel("Rating")
plot.show()
```

### Overall summary of the DataFrame

```python
summary = {
    "total_products": len(indiamart_df),
    "unique_suppliers": indiamart_df["supplier_name"].nunique(),
    "avg_price": indiamart_df["price_in_rupees"].replace(",","",regex=True).astype(int).mean(),
    "median_price": indiamart_df["price_in_rupees"].replace(",","",regex=True).astype(int).median(),
    "avg_rating": indiamart_df["rating"].mean(),
    "gst_verified_pct": indiamart_df["gst_verified"].mean() * 100
}

summary
```
